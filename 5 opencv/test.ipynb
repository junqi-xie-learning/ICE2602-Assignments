# Lab 5: OpenCV

```python
import cv2
import numpy as np
from matplotlib import pyplot as plt
```

```python
import utility
```

## Load Images

```python
files = ['img1.jpg', 'img2.jpg', 'img3.jpg']
```

```python
# Load color images

imgs_rgb = []
for filename in files:
    img_bgr = cv2.imread('images/{}'.format(filename))
    img_rgb = cv2.cvtColor(img_bgr, cv2.COLOR_BGR2RGB)
    imgs_rgb.append(img_rgb.copy())

for img_rgb in imgs_rgb:
    plt.imshow(img_rgb)
    plt.show()
```

```python
# Load grayscale images

imgs_gray = []
for filename in files:
    img_gray = cv2.imread('images/{}'.format(filename), cv2.IMREAD_GRAYSCALE)
    imgs_gray.append(img_gray.copy())

for img_gray in imgs_gray:
    plt.imshow(img_gray, cmap="gray")
    plt.show()
```

## Plot Color Histograms

```python
for i, img in enumerate(imgs_rgb):
    x_label = ['Red', 'Green', 'Blue']
    data = utility.color_histogram(img)

    plt.bar(x_label, data, 1, color='rgb')
    for x, y in zip(x_label, data):
        plt.text(x, y + 0.02, '{:.2f}'.format(y), horizontalalignment='center')

    plt.savefig('histograms/color_hist_{}.png'.format(i + 1))
    plt.show()
```

## Plot Grayscale Histograms

```python
for i, img in enumerate(imgs_gray):
    data = utility.grayscale_histogram(img)
    plt.bar(range(256), data, 1)
    plt.savefig('histograms/grayscale_hist_{}.png'.format(i + 1))
    plt.show()
```

## Plot Gradient Histograms

```python
for i, img in enumerate(imgs_gray):
    data = utility.gradient_histogram(img)
    plt.bar(range(361), data, 1)
    plt.savefig('histograms/gradient_hist_{}.png'.format(i + 1))
    plt.show()
```
